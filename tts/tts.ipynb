```python
import requests
import shutil
import IPython.display as ipd
```

```python
text_to_voice = "Привіт світе"
language = "uk"
```

```python
response = requests.post('http://localhost:8889/synthesize', json={'text': text_to_voice, 'language': language}, stream=True)
```

```python
with open('sample.wav', 'wb') as out_file:
    shutil.copyfileobj(response.raw, out_file)
```

```python
ipd.Audio(filename="sample.wav")
```
